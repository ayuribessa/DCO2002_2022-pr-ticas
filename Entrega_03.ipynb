```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wv 
import sounddevice as sd

violin_sound_file = './Violino.wav'
flute_sound_file = './Flauta.wav'

fa_violin, violin_samples = wv.read(violin_sound_file)
fa_flute, flute_samples = wv.read(flute_sound_file)

fft_size = 1000

yfft_violin = np.fft.fft(violin_samples,fft_size)/fft_size                              
freq_violin = np.arange(0,fa_violin/2,fa_violin/fft_size)           
yfftuni_violin = yfft_violin[0:fft_size//2]                         

yfft_flute = np.fft.fft(flute_samples,fft_size)/fft_size            
freq_flute = np.arange(0,fa_flute/2,fa_flute/fft_size)              
yfftuni_flute = yfft_flute[0:fft_size//2]                            

fig, (ax1, ax2) = plt.subplots(2)
ax1.stem(freq_violin,abs(yfftuni_violin))                            
ax1.set_title('Amplitude unilateral do Espectro de m(t) do Violino') 
ax1.set_xlabel('Frequencia (Hz)')                               
ax1.set_ylabel('|M(f)|')                                      
ax1.axis([0,1500,-10, 5000])                                      

ax2.stem(freq_flute,abs(yfftuni_flute))                             
ax2.set_title('Amplitude unilateral do Espectro de m(t) da flauta') 
ax2.set_xlabel('Frequencia (Hz)')                              
ax2.set_ylabel('|M(f)|')                                       
ax2.axis([0,1500,-10, 5000])                                   
plt.subplots_adjust(hspace = 0.7, right=3,top=2)
plt.show()
```
